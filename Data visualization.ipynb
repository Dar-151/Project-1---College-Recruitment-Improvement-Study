```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
datafile = "cleaned_datasetsss.csv"
df = pd.read_csv(datafile) 
```

### Question 1: What were the most common influences that impacted students decision to attend CCM?

####  This is the code for influence for student decision I decided to make a bar because it made the most sense. Also I didn't know how to count some impact so I counted it toward impact but if I found a way to seperate those cost, location, programs wouldn't be as tight.


```python
# Question 1
impact_columns = [
    'impact_cost', 'impact_location', 'impact_programs', 'impact_online',
    'impact_referral', 'impact_faculty', 'impact_reputation',
    'impact_class_size', 'impact_transfer', 'impact_negative_experience'
]

# Clean up the responses
for col in impact_columns:
    df[col] = df[col].replace({
        'High Impact': 'Impact',
        'Some Impact': 'Impact',
        'No Impact': 'No Impact'
    })

# Count how many said Impact
impact_counts = []
for col in impact_columns:
    count = (df[col] == 'Impact').sum()
    impact_counts.append(count)

# Make bar chart
plt.bar(impact_columns, impact_counts)
plt.xticks(rotation=45, ha='right')
plt.title('Most Common Influences on Students Decision to Attend CCM')
plt.xlabel('Impact Factors')
plt.ylabel('Number of Students')
plt.show()

```

### Question 2: What percentage of survey respondents who attended instant decision day were male?

#### For this question I because I wanted I percentage answer I couldn't do much so I didn;t do a chart I just calculated it. 

```python
# Total number of students who said Yes
total_attended = (df['event_instant_decision'] == 'Yes').sum()

# Number of males who said Yes
male_attended = ((df['event_instant_decision'] == 'Yes') & (df['gender'] == 'Man')).sum()

# Calculate percentage
percent_male = (male_attended / total_attended) * 100

# I only used print because it looked wierd without percent sign
print(percent_male, "%")


```

### Question 3: Which factors had the greatest impact on decison to enroll in Computer Science at CCM? (The fixed version from your comments on my submission)

### I chose the bar chart as well it works for the best for this situation because I don't know how else to show the greatest factors.

```python
# Question 3
impact_columns = [
    'impact_cost', 'impact_location', 'impact_programs', 'impact_online',
    'impact_referral', 'impact_faculty', 'impact_reputation',
    'impact_class_size', 'impact_transfer', 'impact_negative_experience'
]

# Count "Impact"
impact_counts = []
for col in impact_columns:
    count = (df[col] == 'Impact').sum()
    impact_counts.append(count)

#I felt my code was getting repetitive so I wanted to calculate top 5 instead of all factors
# I did use AI for some some of this block
top = sorted(zip(impact_columns, impact_counts), key=lambda x: x[1], reverse=True)[:5]
labels = [x[0] for x in top]
values = [x[1] for x in top]

# Bar chart
plt.bar(labels, values)
plt.xticks(rotation=45, ha='right')
plt.title('Top Factors Influencing Students Decision to Choose Their Major')
plt.xlabel('Impact Factors')
plt.ylabel('Number of Students')
plt.show()
```

### Question 4: Which CCM events had the most participation?

#### I chose this bar because it makes the most sense for this scenario again I think I could've a pie chart but I felt bar made more sense.

```python
# Question 4
event_columns = ['event_open_house', 'event_instant_decision', 'event_info session on campus',
       'event_titans tuesday', 'event_women who dare', 'event_college fair',
       'event_HS sharetime', 'event_challenger program',
       'event_cybersecruity program', 'event_info session from school',
       'event_high school visit', 'event_individual visit',
       'event_workforce development', 'event_cooperate training',
       'event_teen arts festivial', 'event_summer camp at CCM']

# Count how many attended each event
event_counts = []
for col in event_columns:
    count = (df[col] == 'Yes').sum()
    event_counts.append(count)

# Bar chart
plt.bar(event_columns, event_counts, color='skyblue', edgecolor='black')
plt.title('CCM Events with the Most Participation')
plt.xlabel('Event Name')
plt.ylabel('Number of Students')

# FIX → rotate and space labels
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()

```
